#GeV Detectability of SNR G015.4+00.1 

SNR G015.4+00.1 is an interesting SNR that has a TeV detection (HESS J1818−154) and possibly a GeV detection.



#Radio, X-ray, TeV
There's a CO cloud in the north of the remnant and claims that the TeV sources arise from hadronic emission from the interaction of the SNR shell in the cloud (http://arxiv.org/abs/1309.6026v1). 

```python
from IPython.core.display import Image 
Image(filename='/Users/jmcohen/Desktop/G015.4_runptlk/G15_radio_CO.png') 
```

However in http://arxiv.org/abs/1310.6956v2, the TeV sources is shown to be smaller than the rario SNR and containted within it. XMM observations show a diffuse non-thermal x-ray "core" coincident with the TeV source, and a fainter halo that fills the SNR cavity completely. They argue that the TeV emission doesn't come from the entirety of the cloud (some of the cloud lies outside the SNR shell). Since there's no TeV emission with the northern clump, they disfavor the hadronic scenario. No radio or x-ray (nor GeV I assume) pulsations (i.e. pulsars) have been detected from any sources inside the remnant.

##Hadronic Gamma-Ray Flux Estimate

From http://adsabs.harvard.edu/cgi-bin/nph-data_query?bibcode=1994A%26A...287..959D&link_type=ABSTRACT
\begin{equation}F_{\gamma}(>100 MeV) \approx 4.4\times 10^{-7}\theta E_{51} d_{kpc} n_0 ~ph~ cm^{-2}~ s^{-1}\end{equation}

```python
def pi0Flux(t,E,n,d):
    return 4.4e-7*t*E*n/d**2 #ph cm^-2 s^-1
#For ambient density n0 = 1, E_sn = 1e51, effeciency of KE into CRs theta = 0.1, d = 4.2kpc
print pi0Flux(0.1,1,1,4.2),'ph cm^-2 s^-1'
```
